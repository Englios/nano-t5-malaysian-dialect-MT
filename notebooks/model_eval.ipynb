```python
import torch
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

# Load model
MODEL_PATH = "../outputs"
device = "cuda" if torch.cuda.is_available() else "cpu"

tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH)
model = AutoModelForSeq2SeqLM.from_pretrained(MODEL_PATH)
model = model.to(device)
model.eval()
print("Model loaded!")
```

```python
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)

def translate(text, max_new_tokens=64, num_beams=4):
    enc = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        max_length=512, 
        padding=False,
    ).to(device)
    
    enc = {k: v.to(device) for k, v in enc.items() if k != "token_type_ids"}

    with torch.no_grad():
        outputs = model.generate(
            **enc,
            max_length=64,
            do_sample=True,
            top_p=0.9,
            temperature=0.7
        )


    return tokenizer.decode(outputs[0], skip_special_tokens=True)

tests = [
    f"{prefix} Saya suka makan nasi"
    for prefix in [
        "terjemah melayu ke johor:",
        "terjemah melayu ke kelantan:",
        "terjemah melayu ke sabah:",
        "terjemah melayu ke sarawak:",
        "terjemah melayu ke kedah:",
        "terjemah melayu ke pahang:",
        "terjemah melayu ke perak:",
        "terjemah melayu ke terengganu:",
        "terjemah melayu ke melaka:",
        "terjemah melayu ke negeri sembilan:",
        "terjemah melayu ke Melayu:",
        "terjemah melayu ke Inggeris:",
        "terjemah melayu ke Manglish:",
    ]
]

for t in tests:
    out = translate(t)
    print("\nINPUT :", t)
    print("OUTPUT:", out)
    print("-" * 60)
```
